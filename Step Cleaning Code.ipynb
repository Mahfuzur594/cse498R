```python
import pandas as pd

df = pd.read_csv("pre_departure_ml_ready.csv")
```

```python
country_cols = [col for col in df.columns if "Destination Country" in col]
country_cols
```

```python
# Find all USA-related columns dynamically
usa_cols = [
    col for col in df.columns
    if "Destination Country" in col and "usa" in col.lower()
]

usa_cols
```

```python
['Destination Country_Usa', 'Destination Country_usa']
```

```python
df["Destination Country_United States"] = df[usa_cols].max(axis=1)
```

```python
df.drop(columns=usa_cols, inplace=True)
```

```python
if "Destination Country_lithunia" in df.columns:
    df["Destination Country_Lithuania"] = df["Destination Country_lithunia"]
    df.drop(columns=["Destination Country_lithunia"], inplace=True)
```
